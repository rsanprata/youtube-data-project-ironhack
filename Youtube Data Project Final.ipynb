The search() function from the API has a total of 50 results.

Although I can query all JunsKitchen's videos in just one, I'll make a function to query more if I need it in the future.

This function will search the **upload id**, create an empty list of videos, and an object to store the **nextPageToken**, with is what makes the search continue after 50 videos.

Then, it iterates on the uploads response, storing video information and using the token, until the token is 'None', which means we have no more pages to search.

At last, it will return all information from the channel's videos.

```python
def get_channel_videos(channel_id):
    res = youtube.channels().list(id=channel_id,
                                  part='contentDetails').execute()
    playlist_id = res['items'][0]['contentDetails']['relatedPlaylists']['uploads']
    
    videos = []
    nest_page_token = None
    
    while 1:
        res = youtube.playlistItems().list(playlistId=playlist_id,
                                           part='snippet',
                                           maxResults=50,
                                           pageToken=nest_page_token).execute( )
        videos += res['items']
        next_page_token = res.get('nextPageToken')
        
        if next_page_token is None:
            break
    return videos
```

```python
videos = get_channel_videos('UCRxAgfYexGLlu1WHGIMUDqw')
```

Again, I'll make a function to iterate on the results.

First, creating a list with all the videos id's, using a lambda function.

```python
video_ids = list(map(lambda x:x['snippet']['resourceId']['videoId'], videos))
```

Now a function to iterate on the results, storing in a list named 'stats'

```python
def get_videos_stats(video_ids):
    stats = []
    for i in range(0, len(video_ids), 50):
        res = youtube.videos().list(id=','.join(video_ids[i:i+50]),
                                   part='statistics').execute()
        stats += res['items']
        
    return stats
```
